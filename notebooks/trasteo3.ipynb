```python
from statsbombpy import sb
import pandas as pd
competiciones= sb.competitions()
competiciones.head()
```

```python
laliga= competiciones[competiciones['competition_name']== 'La Liga']
laliga
```

```python
partidos_extremos = []
barca_nombre = "Barcelona"
for _, row in laliga.iterrows():
    season_id = row['season_id']
    matches = sb.matches(competition_id=row['competition_id'], season_id=season_id)

    barca_matches = matches[
        (matches['home_team'] == barca_nombre) | (matches['away_team'] == barca_nombre)
    ]

    for _, match in barca_matches.iterrows():
        barca_home = match['home_team'] == barca_nombre
        barca_away = match['away_team'] == barca_nombre

        home_goals = match['home_score']
        away_goals = match['away_score']

        goal_difference = abs(home_goals - away_goals)

        if (barca_home and home_goals - away_goals >= 3) or (barca_away and away_goals - home_goals >= 3):
            partidos_extremos.append(match)
        elif (barca_home and away_goals - home_goals >= 3) or (barca_away and home_goals - away_goals >= 3):
            partidos_extremos.append(match)  
        elif barca_home and home_goals < away_goals:
            partidos_extremos.append(match)  


df_extremos = pd.DataFrame(partidos_extremos)[[
    'match_id', 'match_date', 'kick_off', 'competition', 'season',
    'home_team', 'away_team', 'home_score', 'away_score', 'match_week',
    'competition_stage', 'stadium', 'referee', 'home_managers', 'away_managers'
]]


```
