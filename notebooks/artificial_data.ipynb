GENERAR POSESIONES ARTIFICIALES

```python
def generar_posesion(H, num_pases=10, campo_dim=(120, 80)):
    f = FBM(n=num_pases - 1, hurst=H, length=campo_dim[0], method='daviesharte')
    x = f.fbm()
    y = FBM(n=num_pases - 1, hurst=H, length=campo_dim[1], method='daviesharte').fbm()

    # Normalizar coordenadas
    x = (x - np.min(x)) / (np.max(x) - np.min(x)) * campo_dim[0]
    y = (y - np.min(y)) / (np.max(y) - np.min(y)) * campo_dim[1]

    acciones = np.random.choice(['Pass', 'Carry'], size=num_pases - 1, p=[0.7, 0.3])
    
    tiempos = []
    for accion in acciones:
        if accion == 'Pass':
            t = min(np.random.exponential(scale=1.5), 5)
        else:
            t = min(np.random.exponential(scale=5), 20)
        tiempos.append(t)
    tiempos = np.cumsum(tiempos)

    # Diferencias iniciales
    delta_x = np.diff(x)
    delta_y = np.diff(y)

    max_dist_pase = 40
    max_dist_carry = 15

    escalados_x = [0]
    escalados_y = [0]

    for i, accion in enumerate(acciones):
        dx = delta_x[i]
        dy = delta_y[i]
        dist = np.sqrt(dx**2 + dy**2)
        max_dist = max_dist_pase if accion == 'Pass' else max_dist_carry
        factor = min(1.0, max_dist / dist) if dist > 0 else 1.0

        # Evitar que las conducciones vayan hacia atrás
        if accion == 'Carry' and dx < 0:
            dx *= -1  # Invertimos dirección en x para que no vaya hacia atrás

        dx_scaled = dx * factor
        dy_scaled = dy * factor

        escalados_x.append(escalados_x[-1] + dx_scaled)
        escalados_y.append(escalados_y[-1] + dy_scaled)

    # Iniciar desde una zona razonable del campo
    x_start = np.random.uniform(20, 50)
    y_start = np.random.uniform(20, 60)
    x_final = np.clip(np.array(escalados_x) + x_start, 0, campo_dim[0])
    y_final = np.clip(np.array(escalados_y) + y_start, 0, campo_dim[1])

    num_pases = min(len(x_final), len(y_final), len(tiempos))
    return pd.DataFrame({
        'H': [H] * num_pases,
        'x': x_final[:num_pases],
        'y': y_final[:num_pases],
        't': tiempos[:num_pases],
        'accion': acciones
    })
```

## CREACION DEL MODELO MACHINE LEARNING 

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Cargar datos de posesiones sintéticas
posesiones = pd.read_csv("posesiones_sinteticas.csv")

import numpy as np
import pandas as pd

# Crear el diccionario con listas vacías
dic = {
    'posesion': [], 'H': [], 'num_pases': [], 'num_carries': [],
    'var_x': [], 'var_y': [], 'duracion': [], 'distancia_total': [],
    'velocidad_media': [], 'distancia_por_pase': [], 'angulo_movimiento': [],
    'velocidad_cambio': [], 'aceleracion': []
}

# Función para calcular el ángulo entre dos vectores
def calcular_angulo(x1, y1, x2, y2):
    dx1, dy1 = x2 - x1, y2 - y1
    return np.arctan2(dy1, dx1)

# Agrupar por posesion_id y calcular estadísticas
for posesion_id, grupo in posesiones.groupby("posesion_id"):
    # Extraer datos
    x = grupo["x"].values
    y = grupo["y"].values
    t = grupo["t"].values
    acciones = grupo["accion"].values
    
    # Calcular estadísticas
    num_pases = np.sum(acciones == "Pass")
    num_carries = np.sum(acciones == "Carry")
    var_x = np.mean(np.abs(np.diff(x)))  # Variación media en X
    var_y = np.mean(np.abs(np.diff(y)))  # Variación media en Y
    duracion = t[-1] - t[0]  # Duración total de la posesión
    distancia_total = np.sum(np.sqrt(np.diff(x) ** 2 + np.diff(y) ** 2))  # Distancia recorrida
    velocidad_media = distancia_total / duracion if duracion > 0 else 0  # Velocidad media
    
    # Variables adicionales
    distancia_por_pase = np.sqrt(np.diff(x) ** 2 + np.diff(y) ** 2)  # Distancia entre cada par de puntos consecutivos
    angulo_movimiento = [calcular_angulo(x[i], y[i], x[i+1], y[i+1]) for i in range(len(x)-1)]  # Ángulo de movimiento entre pases
    
    # Calcular cambio de velocidad (diferencia de distancia entre pasos consecutivos / diferencia de tiempo)
    velocidad_cambio = [np.sqrt((x[i+1] - x[i])**2 + (y[i+1] - y[i])**2) / (t[i+1] - t[i]) 
                        for i in range(len(x)-1)] if len(x) > 1 else [0]
    
    # Calcular aceleración
    aceleracion = [velocidad_cambio[i+1] - velocidad_cambio[i] for i in range(len(velocidad_cambio)-1)] if len(velocidad_cambio) > 1 else [0]
    
    # Guardar en el diccionario
    dic["posesion"].append(list(zip(x, y, t)))  # Guardamos la posesión como lista de tuplas
    dic["H"].append(grupo["H"].iloc[0])  # Valor de H
    dic["num_pases"].append(num_pases)
    dic["num_carries"].append(num_carries)
    dic["var_x"].append(var_x)
    dic["var_y"].append(var_y)
    dic["duracion"].append(duracion)
    dic["distancia_total"].append(distancia_total)
    dic["velocidad_media"].append(velocidad_media)
    dic["distancia_por_pase"].append(np.mean(distancia_por_pase) if len(distancia_por_pase) > 0 else 0)  # Promedio de la distancia por pase
    dic["angulo_movimiento"].append(np.mean(angulo_movimiento) if len(angulo_movimiento) > 0 else 0)  # Promedio de los ángulos de movimiento
    dic["velocidad_cambio"].append(np.mean(velocidad_cambio) if len(velocidad_cambio) > 0 else 0)  # Promedio del cambio de velocidad
    dic["aceleracion"].append(np.mean(aceleracion) if len(aceleracion) > 0 else 0)  # Promedio de la aceleración

# Convertir a DataFrame
df = pd.DataFrame(dic)
df.head()


```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from torch.utils.data import DataLoader, TensorDataset

# Configuración de la GPU (si está disponible)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Prepara los datos
X = df.drop(columns=['H', 'posesion']).values  # Características (eliminamos "H" y "posesion")
y = df['H'].values  # Variable objetivo

# Convertir los datos a tensores de PyTorch
X_tensor = torch.tensor(X, dtype=torch.float32).to(device)
y_tensor = torch.tensor(y.values, dtype=torch.float32).to(device)

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)

# Crear DataLoader para manejar el entrenamiento en lotes
train_data = TensorDataset(X_train, y_train)
test_data = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

# Definir la arquitectura del modelo MLP
class MLP(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)  # Capa de entrada a la capa oculta
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)  # Capa oculta a capa oculta
        self.fc3 = nn.Linear(hidden_dim, output_dim)  # Capa oculta a capa de salida
        self.relu = nn.ReLU()  # Función de activación ReLU

    def forward(self, x):
        x = self.relu(self.fc1(x))  # Capa 1 + activación
        x = self.relu(self.fc2(x))  # Capa 2 + activación
        x = self.fc3(x)  # Capa de salida
        return x

# Instanciar el modelo
input_dim = X_train.shape[1]  # Número de características
hidden_dim = 128  # Número de neuronas en la capa oculta
output_dim = 1  # Solo una salida, H

model = MLP(input_dim, hidden_dim, output_dim).to(device)

# Definir la función de pérdida y el optimizador
criterion = nn.MSELoss()  # Error cuadrático medio
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Optimizador Adam

# Función para entrenar el modelo
def train(model, train_loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    for inputs, targets in train_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        
        optimizer.zero_grad()  # Limpiar los gradientes
        outputs = model(inputs)  # Hacer una predicción
        loss = criterion(outputs.flatten(), targets)  # Calcular la pérdida
        loss.backward()  # Retropropagación
        optimizer.step()  # Actualizar los parámetros
        
        running_loss += loss.item()
    return running_loss / len(train_loader)

# Función para evaluar el modelo
def evaluate(model, test_loader, criterion, device):
    model.eval()
    y_true = []
    y_pred = []
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            y_true.extend(targets.cpu().numpy())
            y_pred.extend(outputs.cpu().numpy())
    
    mae = mean_absolute_error(y_true, y_pred)
    return mae

# Entrenamiento y evaluación
num_epochs = 50
for epoch in range(num_epochs):
    train_loss = train(model, train_loader, criterion, optimizer, device)
    mae = evaluate(model, test_loader, criterion, device)
    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {train_loss:.4f}, MAE: {mae:.4f}")

# Evaluar el modelo final
final_mae = evaluate(model, test_loader, criterion, device)
print(f"Final Mean Absolute Error: {final_mae:.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from torch.utils.data import DataLoader, TensorDataset

# Configuración de la GPU (si está disponible)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Prepara los datos
X = df.drop(columns=['H', 'posesion']).values  # Características (eliminamos "H" y "posesion")
y = df['H'].values  # Variable objetivo

# Convertir los datos a tensores de PyTorch
X_tensor = torch.tensor(X, dtype=torch.float32).to(device)
y_tensor = torch.tensor(y.values, dtype=torch.float32).to(device)

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)

# Crear DataLoader para manejar el entrenamiento en lotes
train_data = TensorDataset(X_train, y_train)
test_data = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

# Definir la arquitectura del modelo LSTM
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.relu = nn.ReLU()

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        out = self.fc(lstm_out[:, -1, :])  # Usamos la salida final de la secuencia
        return out

# Instanciar el modelo
input_dim = X_train.shape[1]  # Número de características
hidden_dim = 128  # Número de neuronas en la capa oculta
output_dim = 1  # Solo una salida, H

model_lstm = LSTMModel(input_dim, hidden_dim, output_dim).to(device)

# Definir la función de pérdida y el optimizador
criterion = nn.MSELoss()  # Error cuadrático medio
optimizer = optim.Adam(model_lstm.parameters(), lr=0.001)  # Optimizador Adam

# Función para entrenar el modelo
def train(model, train_loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    for inputs, targets in train_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        
        optimizer.zero_grad()  # Limpiar los gradientes
        outputs = model(inputs.unsqueeze(1))  # Hacer una predicción (LSTM requiere 3D: [batch, seq, features])
        loss = criterion(outputs.flatten(), targets)  # Calcular la pérdida
        loss.backward()  # Retropropagación
        optimizer.step()  # Actualizar los parámetros
        
        running_loss += loss.item()
    return running_loss / len(train_loader)

# Función para evaluar el modelo
def evaluate(model, test_loader, criterion, device):
    model.eval()
    y_true = []
    y_pred = []
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs.unsqueeze(1))
            y_true.extend(targets.cpu().numpy())
            y_pred.extend(outputs.cpu().numpy())
    
    mae = mean_absolute_error(y_true, y_pred)
    return mae

# Entrenamiento y evaluación
num_epochs = 50
for epoch in range(num_epochs):
    train_loss = train(model_lstm, train_loader, criterion, optimizer, device)
    mae = evaluate(model_lstm, test_loader, criterion, device)
    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {train_loss:.4f}, MAE: {mae:.4f}")

# Evaluar el modelo final
final_mae_lstm = evaluate(model_lstm, test_loader, criterion, device)
print(f"Final Mean Absolute Error (LSTM): {final_mae_lstm:.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from torch.utils.data import DataLoader, TensorDataset

# Configuración de la GPU (si está disponible)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Prepara los datos
X = df.drop(columns=['H', 'posesion']).values  # Características (eliminamos "H" y "posesion")
y = df['H'].values  # Variable objetivo

# Convertir los datos a tensores de PyTorch
X_tensor = torch.tensor(X, dtype=torch.float32).to(device)
y_tensor = torch.tensor(y.values, dtype=torch.float32).to(device)

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)

# Crear DataLoader para manejar el entrenamiento en lotes
train_data = TensorDataset(X_train, y_train)
test_data = TensorDataset(X_test, y_test)

train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

# Definir la arquitectura del modelo CNN (1D)
class CNNModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_dim, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(128, 64)
        self.fc2 = nn.Linear(64, output_dim)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.pool(self.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)  # Aplanar el tensor antes de pasarlo a las capas totalmente conectadas
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Instanciar el modelo
input_dim = X_train.shape[1]  # Número de características
output_dim = 1  # Solo una salida, H

model_cnn = CNNModel(input_dim, output_dim).to(device)

# Definir la función de pérdida y el optimizador
criterion = nn.MSELoss()  # Error cuadrático medio
optimizer = optim.Adam(model_cnn.parameters(), lr=0.001)  # Optimizador Adam

# Función para entrenar el modelo
def train(model, train_loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    for inputs, targets in train_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        
        optimizer.zero_grad()  # Limpiar los gradientes
        outputs = model(inputs.unsqueeze(1))  # Hacer una predicción (CNN requiere 3D: [batch, channels, features])
        loss = criterion(outputs.flatten(), targets)  # Calcular la pérdida
        loss.backward()  # Retropropagación
        optimizer.step()  # Actualizar los parámetros
        
        running_loss += loss.item()
    return running_loss / len(train_loader)

# Función para evaluar el modelo
def evaluate(model, test_loader, criterion, device):
    model.eval()
    y_true = []
    y_pred = []
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs.unsqueeze(1))
            y_true.extend(targets.cpu().numpy())
            y_pred.extend(outputs.cpu().numpy())
    
    mae = mean_absolute_error(y_true, y_pred)
    return mae

# Entrenamiento y evaluación
num_epochs = 50
for epoch in range(num_epochs):
    train_loss = train(model_cnn, train_loader, criterion, optimizer, device)
    mae = evaluate(model_cnn, test_loader, criterion, device)
    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {train_loss:.4f}, MAE: {mae:.4f}")


# Evaluar el modelo final
final_mae_cnn = evaluate(model_cnn, test_loader, criterion, device)
print(f"Final Mean Absolute Error (CNN): {final_mae_cnn:.4f}")

```
