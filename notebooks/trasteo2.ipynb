```python
from statsbombpy import sb
import pandas as pd

# 1. Obtener todos los partidos de La Liga
laliga_matches = sb.matches(competition_id=11 , season_id= 90)

# 2. Obtener todos los match_id disponibles con datos 360 (frames)
# Esto puede tardar un poco, se recorre todo el dataset
competitions = sb.competitions()
match_ids_with_frames = []

for _, row in competitions.iterrows():
    matches = sb.matches(row['competition_id'], row['season_id'])
    for match_id in matches['match_id']:
        try:
            frames = sb.frames(match_id=match_id)
            if not frames.empty:
                match_ids_with_frames.append(match_id)
        except:
            # Ignora partidos que no tienen datos 360 o lanzan errores
            continue

# 3. Convertir a set y hacer intersección con los de La Liga
laliga_ids = set(laliga_matches['match_id'])
frames_ids = set(match_ids_with_frames)

# 4. Match IDs que están en ambos
common_match_ids = laliga_ids.intersection(frames_ids)

# 5. (Opcional) Filtrar laliga_matches con solo los que tienen frames
laliga_with_frames = laliga_matches[laliga_matches['match_id'].isin(common_match_ids)]

print(laliga_with_frames[['match_id', 'home_team.home_name', 'away_team.away_name']])
```

```python
from statsbombpy import sb
import pandas as pd

# Lista de tus match_ids
match_ids = frames_ids  # asegúrate de tener esta lista definida

# Descargar competiciones y filtrar solo La Liga
competitions = sb.competitions()

laliga_comps = competitions[competitions['competition_name'] == 'FIFA World Cup']

all_matches = []

# Reunir solo partidos de La Liga
for _, row in laliga_comps.iterrows():
    try:
        matches = sb.matches(competition_id=row['competition_id'], season_id=row['season_id'])
        all_matches.append(matches)
    except:
        continue

# Unir todos los partidos
all_matches_df = pd.concat(all_matches, ignore_index=True)

# Filtrar por tus match_ids
selected_matches = all_matches_df[all_matches_df['match_id'].isin(match_ids)]

# Mostrar match_id, equipos y temporada
result = selected_matches[['match_id', 'home_team', 'away_team', 'season']]
print(result)

```

```python
three= sb.frames(match_id=3788741)
three.head()
import pandas as pd
from statsbombpy import sb

import pandas as pd
from statsbombpy import sb

all_passes = []

for match_id in match_ids_mundial:
    try:
        events = sb.events(match_id=match_id)
        passes = events[events['type'] == 'Pass']
        
        # Filtrar solo los pases exitosos (sin outcome)
        passes = passes[passes['pass.outcome'].isna()]
        
        # Seleccionar columnas
        passes_subset = passes[['id', 'minute', 'second', 'team', 'player',
                                'pass.recipient', 'location', 'pass.end_location']].copy()
        passes_subset['match_id'] = match_id
        
        # Separar coordenadas
        passes_subset[['start_x', 'start_y']] = pd.DataFrame(passes_subset['location'].tolist(), index=passes_subset.index)
        passes_subset[['end_x', 'end_y']] = pd.DataFrame(passes_subset['pass.end_location'].tolist(), index=passes_subset.index)
        passes_subset = passes_subset.drop(columns=['location', 'pass.end_location'])

        all_passes.append(passes_subset)
    except:
        continue

# Combinar en un DataFrame final
all_passes_df = pd.concat(all_passes, ignore_index=True)

# Mostrar resultados
print(all_passes_df.head())
```
